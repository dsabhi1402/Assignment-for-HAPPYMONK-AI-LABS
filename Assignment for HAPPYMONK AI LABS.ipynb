1.Write a program in Python to find the sum of digits of a number using recursion

```python
def sum_of_digits(number):
    if number == 0:
        return 0
    else:
        return number % 10 + sum_of_digits(number // 10)

# example usage
print(sum_of_digits(1234)) # Output: 10
```

3.Python Program to count occurrences of a character in below string and create a dictionary with keys as
letters and values as count of occurrences of that letter.

```python
def count_chars(string):
    """
    This function takes a string as input and returns a dictionary with keys as letters
    and values as the count of occurrences of that letter in the string.
    """
    char_count = {}
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

# Example usage
string = "Hello World"
char_count = count_chars(string)
print(char_count)  # Output: {'H': 1, 'e': 1, 'l': 3, 'o': 2, ' ': 1, 'W': 1, 'r': 1, 'd': 1}
```

C-Write an event based programming that triggers an alert message whenever a camera goes offline

```python
import time

class Camera:
    def __init__(self, id):
        self.id = id
        self.status = "online"
        self.last_seen = time.time()

    def ping(self):
        self.status = "online"
        self.last_seen = time.time()

    def is_offline(self):
        return self.status == "offline"

    def __str__(self):
        return f"Camera {self.id} - {self.status}"

class AlertSystem:
    def __init__(self):
        self.cameras = []

    def register_camera(self, camera):
        self.cameras.append(camera)

    def run(self):
        while True:
            for camera in self.cameras:
                if time.time() - camera.last_seen > 5:
                    camera.status = "offline"
                    if camera.is_offline():
                        print(f"Alert: {camera} has gone offline!")
            time.sleep(1)

if __name__ == "__main__":
    alert_system = AlertSystem()

    camera1 = Camera(1)
    camera2 = Camera(2)

    alert_system.register_camera(camera1)
    alert_system.register_camera(camera2)

    alert_system.run()
```

Write a script encode and decode the json

```python
import json
import urllib.request

# Retrieve JSON data from the URL
url = "https://jsonplaceholder.typicode.com/todos"
response = urllib.request.urlopen(url)
data = response.read().decode()

# Decode the JSON data
decoded_data = json.loads(data)

# Print the decoded data
print(decoded_data)

# Encode the decoded data back to JSON
encoded_data = json.dumps(decoded_data)

# Print the encoded data
print(encoded_data)
```

Find the number of users

```python
import requests

response = requests.get('https://jsonplaceholder.typicode.com/todos')
todos = response.json()

user_ids = set()
for todo in todos:
    user_ids.add(todo['userId'])

num_users = len(user_ids)
print(num_users)  # prints the number of unique user IDs

```

Find number of tasks for each users

```python
import requests


response = requests.get("https://jsonplaceholder.typicode.com/todos")

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the JSON response as a Python list of dictionaries
    todos = response.json()

    # Create a dictionary to store the number of tasks for each user
    user_tasks = {}

    # Loop through the todos and count the number of tasks for each user
    for todo in todos:
        user_id = todo["userId"]
        if user_id not in user_tasks:
            user_tasks[user_id] = 1
        else:
            user_tasks[user_id] += 1

    # Print the results
    for user_id, num_tasks in user_tasks.items():
        print(f"User {user_id} has {num_tasks} tasks.")
else:
    print("Error: could not retrieve tasks.")
```

Find the number of completed and incomplete tasks for each users and rank them

```python
import requests

# Fetch all the Todos
response = requests.get('https://jsonplaceholder.typicode.com/todos')
todos = response.json()

# Group the Todos by user ID
todos_by_user = {}
for todo in todos:
    if todo['userId'] not in todos_by_user:
        todos_by_user[todo['userId']] = []
    todos_by_user[todo['userId']].append(todo)

# Count the number of completed and incomplete tasks for each user
stats = []
for user_id, user_todos in todos_by_user.items():
    completed_tasks = sum(1 for todo in user_todos if todo['completed'])
    incomplete_tasks = len(user_todos) - completed_tasks
    stats.append((user_id, completed_tasks, incomplete_tasks))

# Rank the users based on the total number of completed tasks
ranked_stats = sorted(stats, key=lambda x: x[1], reverse=True)

# Print the results
print("User ID | Completed Tasks | Incomplete Tasks")
print("--------------------------------------------")
for user_id, completed_tasks, incomplete_tasks in ranked_stats:
    print(f"{user_id:7} | {completed_tasks:15} | {incomplete_tasks:16}")
```
